```python
df_density = pd.get_dummies(df, prefix='Density', prefix_sep='', columns=['density'])
df_lat = pd.get_dummies(df_density, prefix='Laterality', prefix_sep='', columns=['laterality'])
df_view = pd.get_dummies(df_lat, prefix='View', prefix_sep='', columns=['view'])
df_final = pd.get_dummies(df_view, prefix='Difficult_Negative_Case', prefix_sep='', columns=['difficult_negative_case'])

print(df_final)
```

```python
print('Original File')
print(df_final.head(10))

df_final.age.fillna(value=df_final.age.mean(), inplace=True)
df_final.BIRADS.fillna(value=0, inplace=True)

print()
print('Missing values replaced')
print(df_final.head(10))
```

```python
df_final.to_csv('modifiedTrain.csv', index=False)
```
