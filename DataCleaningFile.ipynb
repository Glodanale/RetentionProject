```python
df = df.rename(columns={'1st College Semester GPA': '1st_College_Semester_GPA',
                        'High School': 'HighSchool',
                       'Hours Transferred Iin': 'Hours_Transferred_In',
                       'Program Name': 'Program_Name',
                       'Appalachian County': 'Appalachian_County',
                       'TN Geographic Divisions': 'TN_Geographic_Divisions',
                       'Fall 1 Retention': 'Fall_1_Retention',
                       'Spring 1 Retention': 'Spring_1_Retention'})
```

```python
df['1st_College_Semester_GPA'].fillna(value=df['1st_College_Semester_GPA'].mean(), inplace=True)
df['Hours_Transferred_In'].fillna(value=0, inplace=True)
```

```python
df['HOUSING']=df['HOUSING'].replace('Y',1)
```

```python
df['HOUSING'].fillna(value=0, inplace=True)
```

```python
df['ORIENTATION_SESSION']=df['ORIENTATION_SESSION'].replace('D',19)
df['ORIENTATION_SESSION']=df['ORIENTATION_SESSION'].replace('N1',20)
df['ORIENTATION_SESSION']=df['ORIENTATION_SESSION'].replace('OO',21)
df['ORIENTATION_SESSION']=df['ORIENTATION_SESSION'].replace('R',22)
df['ORIENTATION_SESSION']=df['ORIENTATION_SESSION'].replace('R1',23)
df['ORIENTATION_SESSION']=df['ORIENTATION_SESSION'].replace('R2',24)
df['ORIENTATION_SESSION']=df['ORIENTATION_SESSION'].replace('X',25)
```

```python
df['ACT_COMPOSITE'].fillna(value=df['ACT_COMPOSITE'].median(), inplace=True)
```

```python
df['Pell_Eligible']=df['Pell_Eligible'].replace('Y',1)
df['Pell_Eligible']=df['Pell_Eligible'].replace('N',0)
df['Pell_Eligible'].fillna(value=0, inplace=True)
```

```python
df['First_Gen_Status']=df['First_Gen_Status'].replace('Y',1)
df['First_Gen_Status']=df['First_Gen_Status'].replace('N',0)
df['First_Gen_Status'].fillna(value=0, inplace=True)
```

```python
df['Appalachian_County']=df['Appalachian_County'].replace('Yes',1)
df['Appalachian_County'].fillna(value=0, inplace=True)
```

```python
df['TN_Geographic_Divisions'].fillna(value='Middle', inplace=True)
```

```python
df['COLLEGES1'].fillna(value=df['COLLEGE'], inplace=True)
```

```python
df['MAJORS1'].fillna(value=df['MAJOR'], inplace=True)
```

```python
#Create new column to indicate whether someone's college changed

df['College_Change'] = np.where((df['COLLEGE'] == df['COLLEGES1']), 0, 1)
```

```python
#Create new column to indicate whether someone's major changed

df['Major_Change'] = np.where((df['MAJOR'] == df['MAJORS1']), 0, 1)
```

```python
df.drop(['SAT_TOTAL', 'FA_Status_Codes', 'FAFSA', 'DEPT', 'CONCEN', 'Fall_1_Retention', 'COLLEGEF2', 'MAJORF2', 'Fall 2 Retention',
         'COLLEGEF3', 'COLLEGES2', 'Fall 3 Retention', 'MAJORF3', 'MAJORS2', 'Spring 2 Retention', 'Program'], axis=1, inplace=True)
```

```python
df = df[df.CLASS == "FR"]
```

```python
pd.crosstab(df['CLASS'], "Class")
```

```python
df['ORIENTATION_COMPLETED']=df['ORIENTATION_COMPLETED'].replace('Yes',1)
df['ORIENTATION_COMPLETED']=df['ORIENTATION_COMPLETED'].replace('No',0)
df['ORIENTATION_COMPLETED']=df['ORIENTATION_COMPLETED'].replace(3,2)
```

```python
pd.crosstab(df['ORIENTATION_COMPLETED'], "Orientation")
```

```python
df = pd.get_dummies(df, prefix='', prefix_sep='', columns=['GENDER'])
```

```python
df = df.rename(columns={"M": "male", "F": "female"})
df = df.rename(columns={"0": "gender_0","N": "gender_N"})
```

```python
df = pd.get_dummies(df, prefix='college_of_', prefix_sep='', columns=['COLLEGE'])
```

```python
df.isin(['W']).any()
```

```python
df = df[df.HOUSING != "W"]
```

```python
df = df[df.APPDT != "1-Oct-21"]
df = df[df.APPDT != "4-Nov-21"]
df = df[df.APPDT != "24-Jan-22"]
df = df[df.APPDT != "14-Aug-18"]
```

```python
df = df[df.HIGH_SCHOOL_GPA <= 4.0]
```

```python
df = df[df.Hours_Transferred_In <= 120]
```

```python
df.to_csv('cleanData5.csv', index=False)
```
